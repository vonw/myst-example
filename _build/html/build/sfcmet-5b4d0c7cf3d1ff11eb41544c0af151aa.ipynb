```python
import xarray as xr
import hvplot.xarray
import holoviews as hv
from bokeh.plotting import show
```

```python
sfcmet = xr.open_zarr("https://icecaps-o.s3-ext.jc.rl.ac.uk/sfcmet/", consolidated=True)
sfcmet
```

```python
#| label: sfcmetsfcmet['temperature at 2 meters'].sel(time='2017 07').plot(figsize=(12,3))
```

```python
#| label: sfcmet
 
hvfig = sfcmet['temperature at 2 meters'].sel(time=slice('2012 07:08')).hvplot(width=1000, grid=True)
p = hv.render(hvfig)
show(p)
```
